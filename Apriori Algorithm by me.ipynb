## Problem StateMent ==>
Out of the 2000 transctions, 200 contain jam, whereas 300 contain bread. These 300 transactions include a 100 that includes bread and jam.

Using this data, we have to find support , confidence and lift?

## support(l) ==> (transaction conatining)/transactions


## confidence(l1==>l2)==>



## (transaction containing l1 and l2)/transaction



## lift(l1==>l2)==> confidence(l1==>l2)/support(l1)


```python
# (1). Support 
# Support is a default popularity od any item.
# we can calculate the Support as a quotient of the division of the number of transactions containing that item by the total number of transactions.

# Support(jam) = (Transaction involving Jam)/(Total Transactions)
# = 200/200 ==> 10%
```

```python
# (2). confidence ==>
# Confidence is the likelihood that customer bought both bread and jam.
# Dividing the number of transactions that include both bread an jam by the total number of transactions will give the confisdence.

# Confidence = 
# (Traansaction involving bread and jam)/(Total Transaction Involving jam)
# = 100/200 = 50%
# It implies  that 50% of customers who bought jam and bread together.
```

```python
# (3). Lift ==>
# Lift is the increase in the ratio of the sale of bread when we sell jam.

# Lift ==> (Confidence (Jam-Bread)/(support(jam)))

# ==> 50/10 ==> 5

# It Says that the Likelihood of a customer buying both jam and bread together is 5 time more than the chance of purchasing jam alone.
# If the Lift value is less than 1 , it entails that the customer are unlikely to buy the items together.
# Greater the value, the better is the combination.
```
